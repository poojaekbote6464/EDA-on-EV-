```python
import pandas as pd

data = pd.read_csv('dataset.csv')
data.head()
```

```python
data.info()
```

```python
univariate_stats = data.describe()
print(univariate_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

fig, axs = plt.subplots(3, 1, figsize=(10, 15))

sns.histplot(data['Model Year'], bins=30, kde=True, ax=axs[0], color="skyblue")
axs[0].set_title('Distribution of Model Year')

sns.histplot(data['Electric Range'], bins=30, kde=True, ax=axs[1], color="lightgreen")
axs[1].set_title('Distribution of Electric Range')

sns.histplot(data['Base MSRP'], bins=30, kde=True, ax=axs[2], color="salmon")
axs[2].set_title('Distribution of Base MSRP')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(18, 12))

sns.countplot(y=data['Make'], order=data['Make'].value_counts().index, ax=axs[0, 0], palette='Set2')
axs[0, 0].set_title('Count of Vehicles by Make')

sns.countplot(y=data['Model'], order=data['Model'].value_counts().index[:10], ax=axs[0, 1], palette='Set3')
axs[0, 1].set_title('Top 10 Vehicle Models')

sns.countplot(y=data['Electric Vehicle Type'], order=data['Electric Vehicle Type'].value_counts().index, ax=axs[1, 0], palette='viridis')
axs[1, 0].set_title('Count of Vehicles by Electric Vehicle Type')

sns.countplot(y=data['State'], order=data['State'].value_counts().index[:10], ax=axs[1, 1], palette='coolwarm', hue=None)
axs[1, 1].set_title('Count of Vehicles by State')



plt.tight_layout()
plt.show()
```

```python
numeric_data = data.select_dtypes(include=[float, int])

plt.figure(figsize=(10, 6))
correlation_matrix = numeric_data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Electric Range', y='Base MSRP', data=data, hue='Make')
plt.title('Electric Range vs. Base MSRP')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

sns.scatterplot(x='Model Year', y='Electric Range', data=data, ax=axs[0], color="blue")
axs[0].set_title('Model Year vs Electric Range')

sns.scatterplot(x='Model Year', y='Base MSRP', data=data, ax=axs[1], color="green")
axs[1].set_title('Model Year vs Base MSRP')

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

file_path = "dataset.csv"
data = pd.read_csv(file_path)

state_data = data.groupby('State').size().reset_index(name='EV Count')

fig = px.choropleth(state_data,
                    locations='State',  
                    locationmode='USA-states',  
                    color='EV Count', 
                    scope='usa',  
                    color_continuous_scale='green', 
                    title='Number of EV Vehicles by State')

fig.show()
```

```python

ev_count_by_year = data.groupby(['Model Year', 'Make']).size().reset_index(name='Count')

ev_count_by_year = ev_count_by_year.sort_values(by=['Model Year', 'Count'], ascending=[True, False])

ev_count_by_year.head(100)
```

```python
import bar_chart_race as bcr
import pandas as pd

data = pd.read_csv('dataset.csv')

ev_counts_by_year = data.groupby(['Model Year', 'Make']).size().unstack(fill_value=0)
ev_counts_by_year = ev_counts_by_year.sort_index()

ev_counts_by_year.fillna(0, inplace=True)

bcr.bar_chart_race(
    df=ev_counts_by_year,
    filename='ev_make_racing_bar_plot.mp4',
    title='Year wise sales of each Maker from last decade',
    orientation='h',
    sort='desc',
    n_bars=10,
    steps_per_period=45,
    period_length=3000,
    period_label={'x': .95, 'y': .15, 'ha': 'right', 'va': 'center', 'size': 30, 'weight': 'semibold'},  
    bar_kwargs={'alpha': .99, 'lw': 0},  
    writer='ffmpeg',
    figsize=(12, 8),  
)

```
